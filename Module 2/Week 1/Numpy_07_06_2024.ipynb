## Question 10

```python
def maxx(x, y):
    if x >= y:
        return x
    else:
        return y


a = np. array([5, 7, 9, 8, 6, 4, 5])
b = np. array([6, 3, 4, 8, 9, 7, 1])

pair_max = np. vectorize(maxx, otypes=[float])
print(pair_max(a, b))
```

## Question 20

```python
col_sales_index = df.columns.tolist().index('Sales')
data_sales = data[..., col_sales_index]
mean_sales = data_sales.mean()

# Solution 1 (inefficient)
# res = np.full(data_sales.shape, fill_value='Good')
# res[data_sales == mean_sales] = 'Average'
# res[data_sales < mean_sales] = 'Bad'

# Solution 2


@np.vectorize
def categorize_sales(x, mean_sales):
    return 'Good' if x > mean_sales else 'Average' if x == mean_sales else 'Bad'


res = categorize_sales(data_sales, mean_sales)
res[7:10]
```

## Question 21

```python
col_sales_index = df.columns.tolist().index('Sales')
data_sales = data[..., col_sales_index]
mean_sales = data_sales.mean()
data_sales_diff = np.abs(data_sales - mean_sales)
closest_to_mean_sales = data_sales[np.argmin(data_sales_diff)]

res = categorize_sales(data_sales, closest_to_mean_sales)
res[7:10]
```
